```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
def fremdkapital(preis, eigenkapital, kaufkosten=0.078, kreditkosten=0.03, anzahl=2):
    kostengesamt = (preis * (1+kaufkosten)) / anzahl
#     print("kosten gesamt: ", kostengesamt)
    fremdkapital =( kostengesamt - eigenkapital) * (1+kreditkosten)
#     print("fremdkapital: ", fremdkapital)
    return fremdkapital
    
def monatsrate(eigenkapital, fremdkapital, rueckzahlung_in_jahren=30, zinssatz = 0.00750):
    rueckzahlung_in_monaten = rueckzahlung_in_jahren*12
    periodic_zinssatz = (1+zinssatz)**(1/12) - 1
    rate = -1*np.pmt(periodic_zinssatz , rueckzahlung_in_monaten, fremdkapital)
    return rate
```

```python
# LEA
eigen_lea = 330000
eigen_ivan = 100000
zinssatz = 0.0550
preis=1014393
betriebskosten = 124 * 3
jahre = 60

fremd_lea = fremdkapital(preis, eigen_lea)
fremd_ivan = fremdkapital(preis, eigen_ivan)
rate_lea = monatsrate(eigen_lea, fremd_lea, jahre, zinssatz)
rate_ivan = monatsrate(eigen_ivan, fremd_ivan, jahre, zinssatz)
print("LEA")
print("Fremdkapital: ", fremd_lea)
print("Monatliche Tilgung: ", rate_lea)
print("Monatliche Kosten: ", rate_lea + betriebskosten / 2)
print("-----------------")
print("IVAN")
print("Fremdkapital: ", fremd_ivan)
print("Monatliche Tilgung: ", rate_ivan)
print("Monatliche Kosten: ", rate_ivan + betriebskosten / 2)
print("------")
# rate_ivan = monatsrate(eigen_ivan, preis=1014393)
# print("monatsrate ivan: ", rate_ivan)
print("Gesamt:", rate_lea+rate_ivan)
```

```python
def tilgunsplan(rueckzahlung_in_monaten, fremdkapital, zinssatz, monatsrate ):
    periodic_zinssatz = (1+zinssatz)**(1/12) - 1
    principal_remaining = np.zeros(rueckzahlung_in_monaten)
    interest_pay_arr = np.zeros(rueckzahlung_in_monaten)
    principal_pay_arr = np.zeros(rueckzahlung_in_monaten)

    for i in range(0, rueckzahlung_in_monaten):

        if i == 0:
            previous_principal_remaining = fremdkapital
        else:
            previous_principal_remaining = principal_remaining[i-1]

        interest_payment = round(previous_principal_remaining*periodic_zinssatz, 2)
        principal_payment = round(monatsrate - interest_payment, 2)

        if previous_principal_remaining - principal_payment < 0:
            principal_payment = previous_principal_remaining

        interest_pay_arr[i] = interest_payment 
        principal_pay_arr[i] = principal_payment
        principal_remaining[i] = previous_principal_remaining - principal_payment


    month_num = np.arange(rueckzahlung_in_monaten)
    month_num = month_num + 1
    principal_remaining = np.around(principal_remaining, decimals=2)
    rate = interest_pay_arr+principal_pay_arr
    cum_zins = np.cumsum(interest_pay_arr)
    df= pd.DataFrame(data = {"Rate" : rate, 'Zinsen': interest_pay_arr, 'Tilgung': principal_pay_arr, "Rest": principal_remaining, "Kumulative Zinsen" : cum_zins})
    return df
```

```python
fig = make_subplots(
    rows=2, cols=1,
    vertical_spacing=0.03,
    specs=[[{"type": "table"}],
           [{"type": "scatter"}]
          ]
)

fig.add_trace(
        go.Table(
            header=dict(
                    values=['Monat', 'Rate', 'Zinsen', "Tilgung", 'Rest']
                ),
            cells = dict(
                    values =[month_num, rate, interest_pay_arr, principal_pay_arr, principal_remaining]
                )
            ),
        row=1, col=1
    )

fig.add_trace(
        go.Scatter(
                x=month_num,
                y=principal_pay_arr,
                name= "Tilgung"
            ),
        row=2, col=1
    )

fig.append_trace(
        go.Scatter(
            x=month_num, 
            y=interest_pay_arr,
            name="Zinsen"
        ),
        row=2, col=1
    )



fig.update_layout(title='Tilgungsplan',
                   xaxis_title='Monat',
                   yaxis_title='Summe',
                   height= 800,
                   width = 1200,
                   legend= dict(
                           orientation="h",
                           yanchor='top',
                           y=0.47,
                           xanchor= 'left',
                           x= 0.01
                       )
                  )
```
