```python
import numpy as np
import pandas as pd
from quandler import Quandler
from risk import Risk
import matplotlib.pyplot as plt
%load_ext autoreload
%autoreload 2
from  scipy import stats
%matplotlib inline
plt.rcParams["figure.figsize"] = [18, 10]
```

```python
q = Quandler()
r = Risk()
```

```python

df = q.read(['AMZN', 'AAPL', 'REGI', 'GOLD', "ADS"], gte='2001-01-11', lte='2020-08-10', dropna=False)
```

```python
df = q.get_hfi_returns()
```

```python
df.plot()
```

```python
df.head()
```

# VaR & CVaR

1. Historical VaR
2. Parametrics VaR
3. Modified Cornish-Fisher VaR

```python
var_hist = df.aggregate(r.var_historic)
var_gauss=df.aggregate(r.var_gaussian)
var_corfisher = df.aggregate(r.var_cornish_fisher)
```

```python
vars = pd.concat([var_hist,var_gauss,var_corfisher], axis=1)
vars.columns = ["historical", "gaussian", "corfish"]
vars
```

```python
vars.plot.bar()
```
