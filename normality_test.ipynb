```python
import numpy as np
import pandas as pd
from quandler import Quandler
import matplotlib.pyplot as plt
from  scipy import stats
%matplotlib inline
plt.rcParams["figure.figsize"] = [18, 10]
```

```python
q = Quandler()
df = q.read(['AMZN', 'AAPL', 'REGI'], gte='2001-01-11', lte='2020-08-10', dropna=False)
```

```python
df = df.pct_change()
df.head()
```

```python
df["normal"] = np.random.normal(df.iloc[:,1].mean(), df.iloc[:,1].std(), df.shape[0])
```

```python
# a skewness value greater than zero means that there is more weight in the right tail of the distribution.
pd.DataFrame([df.mean(), df.median(), df.std(), df.skew(), df.kurtosis()], index=["mean", "median","std", "skew", "kurtosis"])
```
